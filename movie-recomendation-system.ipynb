```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import CountVectorizer
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
#merging on basic of title
movies = movies.merge(credits,on="title")
```

```python
movies.info()
```

```python
#data cleaning
#task -> add release date in furthur
#what do we need for context based
#genre
#keyword
#title
#overview
#id
#cast
#crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.tail()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#helper function for getting genres tags in a list
def get_tags(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(get_tags)
```

```python
movies['keywords'] = movies['keywords'].apply(get_tags)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
# get top 4 actors
def get_tags_4(obj):
    count = 0
    L = []
    for i in ast.literal_eval(obj):
        if i !=4 :
            count+=1
            L.append(i['character'])
        else :
            break
    return L
        
```

```python
movies['cast'] = movies['cast'].apply(get_tags_4)
```

```python
movies.head()
```

```python
# only get dircetor name from crew
def fetch_dir(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
#now we get overview string into a List format so we can merge with other tags
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python

def convert():
    L = ['overview','genres','keywords','cast']
    for i in L:
        movies[i] = movies[i].apply(lambda x :[j.replace(" ", "") for j in x])
convert()        
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# now convert tags into a string
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.iloc[0].tags
```

```python
from nltk.stem import PorterStemmer

p = PorterStemmer()
```

```python
def stem(data):
    L=[]
    for i in data.split():
        L.append(p.stem(i))
    return " ".join(L)
#stemming 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
vector = CountVectorizer(max_features=5000,stop_words='english')
```

```python
matrix = vector.fit_transform(new_df['tags']).toarray()
```

```python
matrix
```

```python
feature_names = vector.get_feature_names_out()
# Print all feature names
for feature in feature_names:
    print(feature)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
distance_matrix = cosine_similarity(matrix)
```

```python
distance_matrix
#cosine dist of one movie to all
```

```python
sorted(list(enumerate(distance_matrix[1])),reverse = True, key = lambda x : x[1])[1:6] #movie id , distance
```

```python
def recommend_system(movie):
    index = new_df[new_df['title'] == movie].index[0]
    movies_list = sorted(list(enumerate(distance_matrix[index])),reverse = True, key = lambda x : x[1])
    
    for i in movies_list[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_system('Avatar')
```
